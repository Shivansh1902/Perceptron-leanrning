```python
#part of speeech tagging
train_txt=state_union.raw('2005-GWBush.txt')
sample_txt=state_union.raw('2006-GWBush.txt')
custom_sent_tokennizer=PunktSentenceTokenizer(train_txt)
Tokenized=custom_sent_tokennizer.tokenize(sample_txt)
chunkgram="""Chunk: {<RB.?>*<VB.?>*<NNP.?>,+<NN.?>}"""
```

```python
def process_contnt():
  try:
    for i in Tokenized:
      words=nltk.word_tokenize(i)
      tagged=nltk.pos_tag(words)
      chunkParser=nltk.RegexpParser(chunkgram)
      chunk=chunkParser.parse(tagged)
      print(chunk)
      plt.show(chunk.draw())
  except Exception as e:
    print(str(e))
```

```python
import matplotlib
matplotlib.use('Agg')
```

```python
process_contnt()
```

```python
chunk1=r"""Chunk:{<.*>}"""
try:
  for i in Tokenized:
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    chunkParser=nltk.RegexpParser(chunk1)
    chunked=chunkParser.parse(tagged)
    chunked.draw()
    plt.show()
except Exception as e:
  print(str(e))
```
